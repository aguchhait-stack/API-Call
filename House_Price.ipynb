```python
import pandas as pd
import opendatasets as od
pd.set_option('display.max_columns',None)
import warnings
warnings.filterwarnings('ignore')
od.download('https://www.kaggle.com/datasets/dansbecker/melbourne-housing-snapshot')
df=pd.read_csv('/content/melbourne-housing-snapshot/melb_data.csv')
df.head(10)
```

```python
df.info()
```

```python
for x in df.columns:
     print(df[x].value_counts())
```

```python
df['Date']=pd.to_datetime(df['Date'],format='%d/%m/%Y')
```

```python
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
```

```python
df.head()
```

```python
feature=['Rooms','Type','Regionname','Year','Bathroom','YearBuilt','BuildingArea',\
         'Longtitude','Lattitude','Month']
```

```python
X=df[feature]
```

```python
X['YearBuilt']=X['YearBuilt'].fillna(X['YearBuilt'].mean())
X['BuildingArea']=X['BuildingArea'].fillna(X['BuildingArea'].mean())
X['YearBuilt']=X['YearBuilt'].astype(int)
X['Bathroom']=X['Bathroom'].astype(int)
```

```python
X=pd.get_dummies(data=X,columns=['Type','Regionname'])
```

```python
y=df['Price']
```

```python
X.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_leaf_nodes=100,random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,accuracy_score
mse=mean_squared_error(y_test,y_pred)
print(mse)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)
```
